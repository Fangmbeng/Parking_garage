```python
import time
import datetime
import random
from random import choices


class Parking():

    def __init__(self, ticket, price, parking_space=102):
        self.ticket = ticket
        self.parking_space = parking_space
        self.price = price
        self.price = {
            30: 3.99,
            60: 7.99,
            160: 15.99,
            90: 11.99,
            120: 13.99
        }

    def payforticket(self):
        print(self.price)
        request = int(input("Please enter parking time in minutes from list above: "))
        for i in self.price:
            if request == self.price[i]:
                ticket_number = self.takeTicket()
                print(f"Your {ticket_number} price is $ {self.price[i]} ")
                return self.price[i]
            else:
                return "Please input correct value from list" and self.payforticket()

    def takeTicket(self):
        numbers = [str(x) for x in range(50)]
        self.ticket = ''.join(map(lambda ele: random.choice(random.choice(ele)), numbers))
        self.parking_space -= 1
        return self.ticket

    def leaveGarage(self):
        start_time = time.time()
        end_time = time.time() - start_time + (15 * 60)
        intro = self.payforticket()
        if intro is True:
            print(f"Thank you have a nice day."
                  f"Start time: {start_time}"
                  f"End time: {end_time}")
        time.sleep(end_time)
        print(f"Please leave the parking space in 15mins or pay for more parking time to avoid penalty")

    def main():
        price = {
            30: 3.99,
            60: 7.99,
            160: 15.99,
            90: 11.99,
            120: 13.99
        }
        while True:
            parking_space = 1
            if parking_space != 0:
                print(price)
                request = int(input("Please enter parking time in minutes from list above: "))
                if True:
                    if request in price.keys():
                        value = price[request]
                        parking_space -= 1
                        numbers = [str(x) for x in range(50)]
                        ticket_number = ''.join(map(lambda ele: random.choice(random.choice(ele)), numbers))
                        print(f"Your ticket number{ticket_number} \nprice is $ {value}")
                        start_time = datetime.datetime.now()
                        end_time = datetime.timedelta(minutes=request)
                        if True:
                            print(f"Thank you have a nice day\n"
                                  f"Start time: {start_time}\n"
                                  f"End time: {end_time}\n")
                        time.sleep((request - 15) * 60)
                        print(f"Please leave the parking space in 15mins or pay for more parking time to avoid penalty")
                        pay_more = input("Do you wish to continue or quit: ")
                        if pay_more.lower() == "quit":
                            break
                        else:
                            return
                    else:
                        print("Please input correct value from list")
            else:
                print("Sorry no parking space available")
                break

    main()


customer = Parking
print(customer.main())
```
